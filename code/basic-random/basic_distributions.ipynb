```python
%matplotlib inline
# Imports
import numpy
import scipy.stats
import matplotlib.pyplot as plt

import seaborn; seaborn.set()

# Import widget methods
from IPython.html.widgets import *
```

## Uniform distribution

 We can sample a continuous uniform distribution using the ``numpy.random.uniform`` code below.

```python
# Sample random data
r = numpy.random.uniform(0, 1, size=100)
p = plt.hist(r)
```

## Interactive exploration of uniform distribution

```python
def plot_uniform(range_min=0, range_max=1, samples=100):
    # Check assertions
    assert (range_min < range_max)
    
    # Sample random data
    r = numpy.random.uniform(range_min, range_max, samples)
    p = plt.hist(r, normed=True)
    kernel = scipy.stats.gaussian_kde(r)
    r_range = numpy.linspace(min(r), max(r))
    plt.plot(r_range, kernel(r_range))

interact(plot_uniform, range_min=(0, 10),
         range_max = (1, 20),
         samples = (2, 1000))
```

## Normal distribution

```python
# Sample random data
r = numpy.random.normal(0, 1, size=100)
p = plt.hist(r)
```

```python
def plot_normal(mean=0, standard_deviation=1, samples=100, window_range=100):
    # Check assertions
    assert (standard_deviation > 0)
    assert (samples > 1)
    
    # Sample random data
    r = numpy.random.normal(mean, standard_deviation, 
                            size=samples)
    p = plt.hist(r, normed=True)
    kernel = scipy.stats.gaussian_kde(r)
    r_range = numpy.linspace(min(r), max(r))
    plt.plot(r_range, kernel(r_range))
    plt.xlim(min(-window_range, min(r)), max(window_range, max(r)))

interact(plot_normal, mean=(-25, 25),
         standard_deviation = (1, 100),
         samples = (2, 1000),
         window_range = (1, 100))
```

# Poisson Distribution

```python
# Sample random data
r = numpy.random.poisson(5, size=100)
p = plt.hist(r)
```

```python
def plot_poisson(rate=2, samples=100, window_range=20):
    # Check assertions
    assert (rate > 0)
    assert (samples > 1)
    
    # Sample random data
    r = numpy.random.poisson(rate, size=samples)
    f = plt.figure()
    p = plt.hist(r, normed=True)
    kernel = scipy.stats.gaussian_kde(r)
    r_range = numpy.linspace(min(r), max(r))
    plt.plot(r_range, kernel(r_range))
    plt.xlim(-1, max(max(r), window_range))

interact(plot_poisson, rate=(0.1, 100),
         samples = (2, 10000),
         window_range = (1, 100))
```
